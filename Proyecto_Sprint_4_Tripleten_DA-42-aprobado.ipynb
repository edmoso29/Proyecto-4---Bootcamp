### [B1] Diferencia entre miércoles y sábados para  `'order_hour_of_day'`. Traza gráficos de barra para los dos días y describe las diferencias que veas.

```python
miércoles = instacart[instacart['order_dow'] == 3]
sábado = instacart[instacart['order_dow'] == 6]
hora_vs_miércoles = miércoles.groupby('order_hour_of_day')['order_dow'].count()
hora_vs_sábado = sábado.groupby('order_hour_of_day')['order_dow'].count()

miércoles_y_sábado = pd.concat([hora_vs_miércoles, hora_vs_sábado], axis=1)
miércoles_y_sábado.plot(title='Pedidos el miércoles por hora', xlabel='Hora del día', ylabel='Conteo', grid=True, kind='line', figsize=(10, 6))
plt.xticks(range(24))
plt.legend(['Miércoles', 'Sábado'])
```

### [B3] ¿Cuáles son los 20 productos más populares (muestra su ID y nombre)?

```python
popular_products = orders['product_id'].value_counts()
popular_products = pd.DataFrame({'product_id': popular_products.index, 'frequency': popular_products.values})
popular_products_with_names = pd.merge(popular_products, products[['product_id', 'product_name']], on='product_id', how='left')
```

```python
popular_products_with_names['product_id_name'] = popular_products_with_names['product_id'].astype(str) + ' - ' + popular_products_with_names['product_name'].astype(str)
top20_popular_products_names = popular_products_with_names.head(20)
```

```python
top20_popular_products_names.plot(title='Los 20 productos más populares', x='product_id_name', y='frequency', kind='bar', grid=True, figsize=(16, 8))
plt.show()
```

Escribe aquí tus conclusiones.

Los productos más populares son frutas, verduras y leche orgánica saborizada.
